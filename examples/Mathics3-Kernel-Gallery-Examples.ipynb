# Mathics3 Gallery Examples

Here, we have some examples of running Mathics3 in a Jupyter notebook.

## Examples involving calculations


```mathematica
Sin[Pi]
```

```mathematica
(* Calculate Euler's famous equation: *)
E^(Pi I)
```

```mathematica
(* An example using Factorial. Arbitrary-length integers are supported: *)
30!
```

```mathematica
(* A 30-digit numeric approximation of Euler's E: *)
N[E, 30]
(* N[] converts a large integer to a MachineReal without loss of precision. *)
```

```mathematica
(* The first n=8, odd numbers (n+1)**2: *)
n = 8; 2 ^ # & /@ Range[0, n] 
```

```mathematica
N[2^1023] // OutputForm 
```

```mathematica
Sum[2 i + 1, {i, 0, 10}] (* Sum of 1st n odd numbers (n+1)**2 *)
```

```mathematica
n = 8; 2 ^ # & /@ Range[0, n] (* Powers of 2 *)
```

```mathematica
Total[%] (* Sum is 2 ^ n - 1 *)
```

## Functions

Colatz Conjecture. See: [https://oeis.org/A006577](https://oeis.org/A006577).

```mathematica
f[n_] := Module[{a=n, k=0}, While[a!=1, k++; If[EvenQ[a], a=a/2, a=a*3+1]]; k]
```

```mathematica
Table[f[n], {n, 4!}]
```

```mathematica
lindenmayer[i_, b_, h_, j_, r_, n_] := 
  (a = h; p = j; s = k = {}; t = Flatten;
    (Switch[#,
      6, s = {a, p, s},
      8, {a, p, s} = s,
      _C, k = {k, Line@{p, p += {Cos@a, Sin@a}}}];
     If[# < 9, a += I^# b ]) & /@ t@Nest[# /. r &, i, n];
    Graphics@t@k);
lindenmayer[{C[1], X}, Pi/2, 0, {0, 0}, {X -> {X, 4, Y, C[1]}, Y -> {C[1], X, 2, Y}}, 7]
```
